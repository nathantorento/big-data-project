```python
import geopandas as gpd

# # Download the world country shapefiles from Natural Earth (1:110m scale)
# world_shapefile_url = "https://www.naturalearthdata.com/http//www.naturalearthdata.com/download/110m/cultural/ne_110m_admin_0_countries.zip"

# Read the shapefile into a GeoDataFrame
world_gdf = gpd.read_file('ne_110m_admin_0_countries/ne_110m_admin_0_countries.shp')[['ISO_A3', 'geometry']]

# Assuming you have already loaded your 'esgdata' DataFrame
# Merge 'esgdata' with the GeoDataFrame based on the country code or country name
merged_df = pd.merge(world_gdf, esgdata, left_on='ISO_A3', right_on='Country Code')
```

```python
merged_df.head()
```

```python
# Specify the Indicator Name and year
indicator = indicators[0]
year = '2021'
filtered_data = merged_df[merged_df['Indicator Name']==indicator]
filtered_data.head()
```

```python
import matplotlib.pyplot as plt

# Plot the world map using GeoPandas
fig, ax = plt.subplots(figsize=(10, 8))
filtered_data.plot(column=year, cmap='Blues', legend=True, ax=ax)
ax.set_title(indicator)
plt.show()
```
